## 1. Logistic Regression: Regularization

$min J(w,b)$ , where $w \in R^{nx}$, and $b \in R$

$J(w,b) = \frac{1}{m}\sum_{i=1}^{m}L(\hat{Y}^{(i)}, Y^{(i)} ) + \frac{\lambda}{2m}||w||_{2}^{2}$


L2 Regularization $||w||_{2}^{2} = \sum_{j=1}^{nx}w_j^2$
